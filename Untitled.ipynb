```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times"
from scipy import interpolate
def interp2d(d,fx=1,fz=1):                                                                                        
    nx,nz=np.shape(d)                                                                                             
    x = np.linspace(0,nx,nx)                                                                                      
    z = np.linspace(0,nz,nz)                                                                                      
    f = interpolate.interp2d(z,x,  d, kind='cubic')                                                               
    xnew = np.linspace(0,nx,2*nx)                                                                                 
    znew = np.linspace(0,nz,2*nz)                                                                                 
    dnew = f(znew, xnew) #插值后的数据                                                                                  
    return dnew
```
